```python
# give stage 1 and stage 2 elements a floor number
def add_balanced_floor(df, floors=(1, 2), col_name="floor", seed=42):
    n = len(df)
    k = len(floors)
    reps = (n + k - 1) // k  
    seq = (list(floors) * reps)[:n]
    rng = np.random.default_rng(seed)
    rng.shuffle(seq)
    out = df.copy()
    out[col_name] = seq
    return out
```

```python
def expand_tasks(df, allow_window=False):
    rows = []
    for _, r in df.iterrows():
        # 若 df 里没有 'stage' 列，这里会失败；确保外部已给各 df 设置了 stage=1/2/3
        stage = int(r['stage'])
        element = str(r['element'])
        # 更稳妥地取 floor，避免 KeyError
        floor_val = r.get('floor', pd.NA)
        floor = floor_val if pd.notna(floor_val) else 'NA'
        wt_code = r['wt_code']

        # 基础步骤
        steps = BASE_PIPELINE.copy()

        # S3 且 facade：插入窗
        if allow_window and wt_code == 'F':
            steps = ['Element Assembly', 'Window Installation', 'Transportation', 'Facade Work']
        # Roof / Interior
        elif wt_code == 'R':
            steps = ['Element Assembly', 'Transportation', 'Roof Work']
        elif wt_code == 'I':
            steps = ['Element Assembly', 'Transportation', 'Interior Work']

        pre_process = ""
        for step in steps:
            step_code = STEPCODE[step]  # 确保 step 在 STEPCODE 中
            task_id = f"S{stage}_F{floor}_{step_code}_{element}"
            rows.append({
                'task_id': task_id,
                'stage': stage,
                'floor': floor,
                'element': element,
                'task_type': step,          # 完整名称保留
                'work_type_code': wt_code,  # 只存简写
                'pre_process': pre_process  # 元素内部的前置
            })
            pre_process = task_id
    return pd.DataFrame(rows)

tasks_s1 = expand_tasks(stage_1_data, allow_window=False)
tasks_s2 = expand_tasks(stage_2_data, allow_window=False)
tasks_s3 = expand_tasks(stage_3_data, allow_window=True)
```

```python
# —— 预先为“关键路径计算”准备全局的跨任务前后约束（不放到 all_tasks 里）——
# 1) 楼层/阶段顺序：S1_F1 → S1_F2 → S2_F1 → S2_F2（仅指 OI 批次的串联）
def _filter_oi(df, stage, floor):
    return df[(df['stage'] == stage) & (df['floor'].astype(str) == str(floor)) & (df['task_type'] == 'Installation Work')]

edges = []  # (pred_task_id, succ_task_id)

# 假设楼层集合来自 S1/S2 的实际 floor 值（去掉 NA）
floors_order = sorted({f for f in pd.concat([stage_1_data['floor'], stage_2_data['floor']]).dropna().astype(str).unique()},
                      key=lambda x: (x == 'NA', x))  # 简单排序；如需 0<1<2…可改成按 int 排
```

```python
# 串 S1 楼层链
for i in range(len(floors_order)-1):
    fl_from, fl_to = floors_order[i], floors_order[i+1]
    pred_ids = _filter_oi(all_tasks, 1, fl_from)['task_id'].tolist()
    succ_ids = _filter_oi(all_tasks, 1, fl_to)['task_id'].tolist()
    for p in pred_ids:
        for s_ in succ_ids:
            edges.append((p, s_))

# 串 S2 楼层链：S2_F1 → S2_F2
for i in range(len(floors_order)-1):
    fl_from, fl_to = floors_order[i], floors_order[i+1]
    pred_ids = _filter_oi(all_tasks, 2, fl_from)['task_id'].tolist()
    succ_ids = _filter_oi(all_tasks, 2, fl_to)['task_id'].tolist()
    for p in pred_ids:
        for s_ in succ_ids:
            edges.append((p, s_))

# 串阶段链：S1 完成 → S2 同楼层开始
for fl in floors_order:
    s1_oi_ids = _filter_oi(all_tasks, 1, fl)['task_id'].tolist()
    s2_oi_ids = _filter_oi(all_tasks, 2, fl)['task_id'].tolist()
    for p in s1_oi_ids:
        for s_ in s2_oi_ids:
            edges.append((p, s_))

# 2) S3 的 Onsite Installation 需等同楼层 S1&S2 的 OI 都完成
for fl in floors_order:
    s3_oi_ids = _filter_oi(all_tasks, 3, fl)['task_id'].tolist()
    if not s3_oi_ids:
        continue
    s12_oi_ids = _filter_oi(all_tasks, 1, fl)['task_id'].tolist() + _filter_oi(all_tasks, 2, fl)['task_id'].tolist()
    for s3_task in s3_oi_ids:
        for p in s12_oi_ids:
            edges.append((p, s3_task))

# 去重并构造成 DataFrame
precedence_edges = pd.DataFrame(edges, columns=['pred', 'succ']).drop_duplicates()
```

```python
def add_virtual_ends(tasks_df, edges_df,
                     id_col='task_id', duration_col='Duration',
                     end_id='__END__'):
    tasks = tasks_df.copy()
    edges = edges_df[['pred','succ']].astype(str).drop_duplicates().copy()
    nodes = set(tasks[id_col].astype(str))

    succs = set(edges['pred'])
    terminal_nodes = sorted(nodes - succs)  # 无后继 => 终点

    # 1) 加入虚拟 END 节点（工期=0）
    if end_id not in set(tasks[id_col].astype(str)):
        tasks = pd.concat([
            tasks,
            pd.DataFrame([{id_col: end_id, duration_col: 0}])
        ], ignore_index=True)

    # 2) 把所有终点连到 END
    extra = pd.DataFrame({'pred': terminal_nodes, 'succ': end_id})
    edges = pd.concat([edges, extra], ignore_index=True).drop_duplicates()

    return tasks, edges
```

```python
import pandas as pd
from collections import defaultdict, deque

def compute_cpm_with_duration(tasks_df: pd.DataFrame,
                              edges_df: pd.DataFrame,
                              id_col: str = 'task_id',
                              duration_col: str = 'Duration'):
    """
    基于固定工期列（整数或浮点）计算关键路径 CPM
    返回：
      result_df: 含 ES, EF, LS, LF, Slack, IsCritical
      critical_paths: 关键路径列表（可能多条），每条是 task_id 序列
      project_duration: 项目总工期
    """
    # 1) 任务与工期
    df = tasks_df[[id_col, duration_col]].copy()
    df[id_col] = df[id_col].astype(str)
    df['DU'] = pd.to_numeric(df[duration_col], errors='coerce').fillna(0.0)

    # 2) 规范边，构图（允许孤立任务）
    edges = edges_df[['pred','succ']].drop_duplicates().astype(str)
    nodes = set(df[id_col].tolist()) | set(edges['pred']) | set(edges['succ'])
    nodes = sorted(nodes)

    succs = defaultdict(list)
    preds = defaultdict(list)
    in_deg = {n: 0 for n in nodes}
    for p, s in edges[['pred','succ']].itertuples(index=False, name=None):
        succs[p].append(s)
        preds[s].append(p)
        in_deg[s] += 1  # 入度统计

    # 3) 拓扑排序（Kahn）并校验 DAG
    topo, q = [], deque([n for n in nodes if in_deg.get(n, 0) == 0]) # 入度为0的节点,也就是没有前驱的节点，起始点
    while q:
        u = q.popleft()
        topo.append(u)
        for v in succs.get(u, []):
            in_deg[v] -= 1
            if in_deg[v] == 0:
                q.append(v)
    if len(topo) != len(nodes):
        raise ValueError("检测到环（非 DAG）。请检查依赖是否形成闭环。")

    # 4) 工期映射
    du_map = dict(zip(df[id_col], df['DU']))
    for n in nodes:
        du_map.setdefault(n, 0.0)

    # 5) Forward pass
    ES, EF = {}, {}
    for u in topo:
        ES[u] = 0.0 if len(preds.get(u, [])) == 0 else max(EF[p] for p in preds[u])
        EF[u] = ES[u] + du_map[u]
    project_duration = max(EF.values()) if EF else 0.0

    # 6) Backward pass
    LF, LS = {}, {}
    terminals = [u for u in nodes if len(succs.get(u, [])) == 0]
    for u in terminals:
        LF[u] = project_duration
        LS[u] = LF[u] - du_map[u]
    for u in reversed(topo):
        if len(succs.get(u, [])) > 0:
            LF[u] = min(LS[v] for v in succs[u])
            LS[u] = LF[u] - du_map[u]

    # 7) Slack & critical
    Slack = {u: LF[u] - EF[u] for u in nodes}
    IsCritical = {u: abs(Slack[u]) < 1e-9 for u in nodes}

    # 8) 提取关键路径（可能多条）
    starts = [u for u in nodes if len(preds.get(u, [])) == 0]
    ends   = [u for u in nodes if len(succs.get(u, [])) == 0]
    critical_paths = []

    def dfs_path(u, path):
        path.append(u)
        if u in ends:
            critical_paths.append(path.copy())
        else:
            extended = False
            for v in succs.get(u, []):
                if IsCritical[u] and IsCritical[v] and abs(EF[u] - ES[v]) < 1e-9:
                    extended = True
                    dfs_path(v, path)
            if not extended and u in ends:
                critical_paths.append(path.copy())
        path.pop()

    for s in starts:
        if IsCritical[s]:
            dfs_path(s, [])
    if not critical_paths and nodes:
        # 若没有严格意义的关键链（极少见），给一条最长时间链作为兜底
        end_node = max(nodes, key=lambda x: EF[x])
        chain = []
        cur = end_node
        while True:
            chain.append(cur)
            cands = [p for p in preds.get(cur, []) if abs(EF[p] - ES[cur]) < 1e-9]
            if not cands:
                break
            cur = max(cands, key=lambda x: EF[x])
        critical_paths = [list(reversed(chain))]

    # 9) 汇总
    result_df = pd.DataFrame({
        'task_id': nodes,
        'DU': [du_map[n] for n in nodes],
        'ES': [ES[n] for n in nodes],
        'EF': [EF[n] for n in nodes],
        'LS': [LS[n] for n in nodes],
        'LF': [LF[n] for n in nodes],
        'Slack': [Slack[n] for n in nodes],
        'IsCritical': [IsCritical[n] for n in nodes],
    }).sort_values(['ES','EF','task_id']).reset_index(drop=True)

    return result_df, critical_paths, project_duration

```

```python
tasks2, edges2 = add_virtual_ends(all_tasks, edges_all,
                                  id_col='task_id', duration_col='Duration',
                                  end_id='__END__')

# 再跑之前的 CPM 函数
result_df, critical_paths, project_duration = compute_cpm_with_duration(
    tasks_df=tasks2,
    edges_df=edges2,
    id_col='task_id',
    duration_col='Duration'
)
```
