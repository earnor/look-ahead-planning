```python
HOUR = 60.0  # 统一用“分钟”为模拟时间单位；1 小时 = 60 分钟

# —— 工作日历（可改：周一~周五，08:00–12:00、13:00–17:00）——
MINUTES_PER_DAY = 24 * 60
WORKING_WEEKDAYS = {0, 1, 2, 3, 4}   # Mon=0 ... Sun=6
SHIFT_WINDOWS = [(8*60, 12*60), (13*60, 17*60)]  # (start, end) in minutes
```

```python
ASSEMBLY_CREWS = 2
TRUCKS = 2
INSTALL_CREWS = 2
```

```python
# —— 预留的延迟函数（以后可插入工序级别的延误逻辑；单位：分钟）——
def extra_delay_minutes(task_name: str, elem: dict) -> float: # 我之后应该要改成task type, element就是task id
    """
    可按需返回额外延误（分钟）。例如：
    if task_name == "transport" and elem["id"] == "E3": return 90  # 运输堵车 1.5 小时
    """
    return 0.0
```

```python
def is_working_minute(t: float) -> bool:
    day = (t // MINUTES_PER_DAY) % 7
    if day not in WORKING_WEEKDAYS:
        return False
    minute_in_day = t % MINUTES_PER_DAY
    for s, e in SHIFT_WINDOWS:
        if s <= minute_in_day < e:
            return True
    return False
```

```python
def next_working_time(t: float) -> float:
    """若 t 在非工作时段，返回下一次开工时刻；否则返回 t。"""
    if is_working_minute(t):
        return t
    t0 = t
    while True:
        day = (t0 // MINUTES_PER_DAY) % 7
        minute_in_day = t0 % MINUTES_PER_DAY
        if day in WORKING_WEEKDAYS:
            for s, e in SHIFT_WINDOWS:
                s_abs = (t0 - minute_in_day) + s
                e_abs = (t0 - minute_in_day) + e
                if t0 <= s_abs:
                    return s_abs
                if s <= minute_in_day < e:  # 落在当日工作窗内
                    return t0
        # 跳到次日 00:00
        t0 = ((t0 // MINUTES_PER_DAY) + 1) * MINUTES_PER_DAY
```

```python
def work_timeout(env: simpy.Environment, minutes: float):
    """
    只在工作时段内流逝 minutes；遇到下班/周末自动等待到下一工作窗。
    """
    remaining = minutes
    while remaining > 0:
        now = int(env.now)
        start = next_working_time(now)
        if start > now:
            yield env.timeout(start - now)
        # 找出当前工作窗的结束
        minute_in_day = int(env.now) % MINUTES_PER_DAY
        current_end = None
        for s, e in SHIFT_WINDOWS:
            if s <= minute_in_day < e:
                current_end = (int(env.now) - minute_in_day) + e
                break
        if current_end is None:
            # 不在工作窗起点，跳回去再判断
            continue
        can_do = min(remaining, current_end - env.now)
        yield env.timeout(can_do)
        remaining -= can_do
```

```python
class ProjectDES:
    def __init__(self, env, elements, install_sequence):
        self.env = env
        # 资源
        self.assembly = simpy.Resource(env, capacity=ASSEMBLY_CREWS)
        self.trucks = simpy.Resource(env, capacity=TRUCKS)
        self.install = simpy.Resource(env, capacity=INSTALL_CREWS)

        # 安装顺序 gate：只有轮到该 element 才能开始安装
        self.turn_gate = {eid: env.event() for eid in install_sequence}
        if install_sequence:
            self.turn_gate[install_sequence[0]].succeed()  # 首件可立即安装

        # 记录下一个应被放行的 element（在“完成安装”时推进）
        self.install_order = install_sequence
        self.next_index = 0

        # 统计
        self.records = {e["id"]: {
            "assembly_start": None, "assembly_end": None,
            "transport_start": None, "transport_end": None,
            "install_start": None,  "install_end": None
        } for e in elements}

        # 启动所有 element 流程
        self.elements = {e["id"]: e for e in elements}
        for e in elements:
            env.process(self.element_flow(e))

    def element_flow(self, elem: dict):
        eid = elem["id"]

        # ===== 1) Assembly =====
        with self.assembly.request() as req:
            yield req
            self.records[eid]["assembly_start"] = self.env.now
            dur = elem["assembly_hr"] * HOUR + extra_delay_minutes("assembly", elem)
            yield self.env.process(work_timeout(self.env, dur))
            self.records[eid]["assembly_end"] = self.env.now

        # ===== 2) Transport =====
        with self.trucks.request() as req:
            yield req
            self.records[eid]["transport_start"] = self.env.now
            dur = elem["transport_hr"] * HOUR + extra_delay_minutes("transport", elem)
            yield self.env.process(work_timeout(self.env, dur))
            self.records[eid]["transport_end"] = self.env.now

        # ===== 3) Installation（必须：轮到它 & 有安装班组）=====
        # 等待“轮到它”的 turn gate 打开（体现安装顺序）
        if eid in self.turn_gate and not self.turn_gate[eid].triggered:
            yield self.turn_gate[eid]

        with self.install.request() as req:
            yield req
            self.records[eid]["install_start"] = self.env.now
            dur = elem["install_hr"] * HOUR / INSTALL_CREWS + extra_delay_minutes("install", elem) #这里可以加上crew number以改变施工时间
            yield self.env.process(work_timeout(self.env, dur))
            self.records[eid]["install_end"] = self.env.now

        # 安装完成后，放行下一个（按给定 INSTALL_SEQUENCE）
        if self.install_order and self.next_index < len(self.install_order):
            if self.install_order[self.next_index] == eid:
                self.next_index += 1
                if self.next_index < len(self.install_order):
                    nxt = self.install_order[self.next_index]
                    if not self.turn_gate[nxt].triggered:
                        self.turn_gate[nxt].succeed()
```

```python
def format_minute(m):
    """把从仿真起点算起的分钟数 m → 'D天 HH:MM' """
    if m is None:
        return None
    d, r = divmod(int(m), MINUTES_PER_DAY)
    h, r = divmod(r, 60)
    return f"D{d} {h:02d}:{r:02d}"   # 例如 D0 09:30, D2 14:05

def run_sim(elements, install_seq, until_days=60):
    env = simpy.Environment()
    model = ProjectDES(env, elements, install_seq)
    env.run(until=until_days * MINUTES_PER_DAY)

    # 原始记录（分钟）
    raw_records = model.records

    # 做一份“格式化后的”记录
    formatted_records = {}
    for eid, rec in raw_records.items():
        formatted_records[eid] = {
            "assembly_start":  format_minute(rec["assembly_start"]),
            "assembly_end":    format_minute(rec["assembly_end"]),
            "transport_start": format_minute(rec["transport_start"]),
            "transport_end":   format_minute(rec["transport_end"]),
            "install_start":   format_minute(rec["install_start"]),
            "install_end":     format_minute(rec["install_end"]),
        }

    # 也顺便算一下总工期（用原始分钟算，再转成天）
    starts = [r["assembly_start"] for r in raw_records.values() if r["assembly_start"] is not None]
    ends   = [r["install_end"]    for r in raw_records.values() if r["install_end"]    is not None]
    makespan_days = None
    if starts and ends:
        makespan_min = max(ends) - min(starts)
        makespan_days = makespan_min / MINUTES_PER_DAY

    # 想要分钟版就用 raw_records，想要好看版就用 formatted_records
    return model, formatted_records, makespan_days
```

```python
model, records, makespan_days = run_sim(element_input, install_sequence, until_days=60)
```
